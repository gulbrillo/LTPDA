## Load LTPDA test data

```python

ltpda_ts_a = TSData.from_txt_file('/Users/hewitson/Data/pyda/ltpda_ts_a.txt')
ltpda_ts_a.name = 'ltpda_ts_a'
print(ltpda_ts_a)

ltpda_ts_b = TSData.from_txt_file('/Users/hewitson/Data/pyda/ltpda_ts_b.txt')
ltpda_ts_b.name = 'ltpda_ts_b'
print(ltpda_ts_b)

ltpda_ts_d = TSData.from_txt_file('/Users/hewitson/Data/pyda/ltpda_ts_d.txt')
ltpda_ts_d.name = 'ltpda_ts_d'
print(ltpda_ts_d)

ltpda_S_a = FSData.from_txt_file('/Users/hewitson/Data/pyda/ltpda_S_ts_a.txt')
print(ltpda_S_a)

ltpda_S_b = FSData.from_txt_file('/Users/hewitson/Data/pyda/ltpda_S_ts_b.txt')
print(ltpda_S_b)

ltpda_T_ab = FSData.from_complex_txt_file('/Users/hewitson/Data/pyda/ltpda_T_ab.txt')
print(ltpda_T_ab)

ltpda_Cab = FSData.from_complex_txt_file('/Users/hewitson/Data/pyda/ltpda_C_ab.txt')
print(ltpda_Cab)

ltpda_Coh_db = FSData.from_complex_txt_file('/Users/hewitson/Data/pyda/ltpda_Coh_db.txt')
print(ltpda_Coh_db)
```

## Compute PSDs

```python


Sa = psd(ts=ltpda_ts_a, window='hanning', navs=100, percent_overlap=50, scale='PSD', detrend_order=1)
Sb = psd(ts=ltpda_ts_b, window='hanning', navs=100, percent_overlap=50, scale='PSD', detrend_order=1)
print(Sa)
```

## Compute TFE

```python

T = tfe(ts_a=ltpda_ts_a, ts_b=ltpda_ts_b, window='hanning', navs=10, detrend_order=1, percent_overlap=50)
print(T)
```

## Plot Spectra

```python

ltpda_S_a.linewidth = 4
ltpda_S_b.linewidth = 4
Sa.linewidth = 4
Sb.linewidth = 4
Sa.linestyle = '--'
Sb.linestyle = '--'

ltpda_S_a.loglog(ltpda_S_b, Sa, Sb)
```

## Compare TFs

```python
print(T)

ltpda_T_ab.dyaxis=numpy.array([0])
print(ltpda_T_ab)
print(ltpda_T_ab.yaxis.ddata.size)

ltpda_T_ab.loglog(T, ShowErrors=True)

```

## Compute CPSD

```python
Cab = cpsd(ts_a=ltpda_ts_a, ts_b=ltpda_ts_b, window='hanning', navs=10, detrend_order=1, percent_overlap=50)

print(ltpda_Cab)
print(Cab)
```

```python
ltpda_Cab.dyaxis=numpy.array([0])
ltpda_Cab.loglog(Cab, ShowErrors=True)
```

## Compute Coherence from d to b

```python

Coh_db = cohere(ts_a=ltpda_ts_d, ts_b=ltpda_ts_b, window='hanning', navs=10, detrend_order=1, percent_overlap=50)

print(Coh_db)
```

```python

M_ltpda_Coh_db = ltpda_Coh_db.abs()
M_Coh_db = Coh_db.abs()

M_ltpda_Coh_db.semilogx(M_Coh_db, ShowErrors=True, errortype="area")
```

## Mag-square coherence

```python

Cms_db = mscohere(ts_a=ltpda_ts_d, ts_b=ltpda_ts_b, window='hanning', navs=10, detrend_order=1, percent_overlap=50)

print(Coh_db)
```
